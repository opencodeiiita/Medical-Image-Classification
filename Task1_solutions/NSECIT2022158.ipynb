```python
#Create two lists, images and labels. Store Input images in 'Images' list and its labels in 'Labels' list.
#Infected Images have two categories viral and bacterial. Find Number of images in each class. (Hint: See image name)
#Note that images are of different sizes.

Images = []
Labels = []
normalPath = '/content/OpenCodeData/OpenCodeData/NORMAL'
infectedPath = '/content/OpenCodeData/OpenCodeData/INFECTED'
normalList = os.listdir(normalPath)
infectedList = os.listdir(infectedPath)

for obj in normalList:
  image_path = os.path.join(normalPath, obj)
  Labels.append('normal')
  image = Image.open(image_path)
  Images.append(image)

size = (700,700)

for obj in infectedList:
  image_path = os.path.join(infectedPath, obj)
  label_components = obj.split('_')
  Labels.append(label_components[1])
  image = Image.open(image_path)
  image.resize(size)
  Images.append(image)


normal_count = Labels.count("normal")
viral_count = Labels.count("virus")
bacterial_count = Labels.count("bacteria")

print(f"Number of Normal Images: {normal_count}")
print(f"Number of Viral Images: {viral_count}")
print(f"Number of Bacterial Images: {bacterial_count}")
```
