```python
#Create two lists, images and labels. Store Input images in 'Images' list and its labels in 'Labels' list.
#Infected Images have two categories viral and bacterial. Find Number of images in each class. (Hint: See image name)
#Note that images are of different sizes.
images=[]
labels=[]
viralCount=0
bacterialCount=0
normalCount=0

path='/content/drive/MyDrive/OpenCodeData'
infectedPath='/content/drive/MyDrive/OpenCodeData/INFECTED'
normalPath='/content/drive/MyDrive/OpenCodeData/NORMAL'
size=(256,256)

for root, dirs, files in os.walk(path):
  for x in files:
    if x.endswith('jpeg'):
      if 'virus' in x:
        labels.append('Viral')
        viralCount+=1
      elif 'bacteria' in x:
        labels.append('Bacteria')
        bacterialCount+=1
      else:
        labels.append('Normal')
        normalCount+=1
      img=Image.open(os.path.join(root,x))
      resized_img=img.resize(size)
      images.append(resized_img)

infectedCount = viralCount + bacterialCount
totalCount = infectedCount + normalCount
print(f"Number of Viral Images: {viralCount}")
print(f"Number of Bacterial Images: {bacterialCount}")
print(f"Number of Normal Images: {normalCount}")
print(f"Number of Infected Images: {infectedCount}")
print(f"Total Number of Images: {totalCount}")

```
