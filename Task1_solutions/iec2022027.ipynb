```python
import os
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
from PIL import Image
```

```python
#Create two lists, images and labels. Store Input images in 'Images' list and its labels in 'Labels' list.
#Infected Images have two categories viral and bacterial. Find Number of images in each class. (Hint: See image name)
#Note that images are of different sizes.

path = '../OpenCodeData/OpenCodeData/'

images=[]
labels=[]  # Normal=0,Infected_viral=1, Infected_Bacterial=2  



n_viral=0
n_bacterial=0

for folder in os.listdir(path=path):
    for image in os.listdir(path+folder):
        img = Image.open(path+folder+'/'+image)
        imgr = img.resize((500, 500))
        imgr.save(path+folder+'/'+image)
        images.append(path+folder+'/'+image)
        if(folder.lower()=="infected"):
            labels.append(1)
            if "bacteria" in image.lower():
                labels[-1]+=1
                n_bacterial+=1
            else:
                n_viral+=1
        else:
            labels.append(0)

print(f"Total number of viral images are: {n_viral}")
print(f"Total number of bacterial images are: {n_bacterial}")


```
