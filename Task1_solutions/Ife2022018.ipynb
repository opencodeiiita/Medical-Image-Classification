```python
#Create two lists, images and labels. Store Input images in 'Images' list and its labels in 'Labels' list.
#Infected Images have two categories viral and bacterial. Find Number of images in each class. (Hint: See image name)
#Note that images are of different sizes.
images=[]
labels=[]
infected_images=[]
normal_images=[]
Viral_infected=0
Bacteria_infected=0
infected=0
normal=0
normal_images_path="/content/OpenCodeData/OpenCodeData/NORMAL"
infected_images_path="/content/OpenCodeData/OpenCodeData/INFECTED"
##"/content/OpenCodeData/OpenCodeData/INFECTED"

for img in [i for i in os.listdir(infected_images_path) if i.endswith(".jpeg")]:
    image_path = os.path.join(infected_images_path, img)
    image = cv2.resize(plotimg.imread(image_path, 1), (250, 250))
    images.append(image)
    infected_images.append(image)
    if "bacteria" in img:
        labels.append("Bacteria")
        infected+=1
    elif "virus" in img:
        labels.append("Virus")
        infected +=1

for img in [i for i in os.listdir(normal_images_path) if i.endswith(".jpeg")]:
    image_path = os.path.join(normal_images_path, img)
    image = cv2.resize(plotimg.imread(image_path, 1), (250, 250))
    images.append(image)
    normal_images.append(image)
    labels.append("Normal")

Bacteria_infected = labels.count("Bacteria")
Viral_infected = labels.count("Virus")
normal = labels.count("Normal")
print(f"Number of Bacterial Images ={Bacteria_infected}")
print(f"Number of Viral Images ={Viral_infected}")
print(f"Number of Normal Image ={normal}")
print(f"Number of Infected Images ={infected}")
```
