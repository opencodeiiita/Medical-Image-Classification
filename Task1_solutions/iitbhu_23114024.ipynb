```python
#Create two lists, images and labels. Store Input images in 'Images' list and its labels in 'Labels' list.
#Infected Images have two categories viral and bacterial. Find Number of images in each class. (Hint: See image name)
#Note that images are of different sizes.

Images = []
Labels = []

infectedPath = "OpenCodeData/OpenCodeData/INFECTED"
normalPath = "OpenCodeData/OpenCodeData/NORMAL"

normalAmount = len(os.listdir(normalPath))

infectedAmountBacterial = 0
infectedAmountViral = 0

for file in os.listdir(normalPath):
  image= Image.open(os.path.join(normalPath, file)).resize((224, 224))
  Images.append(image)
  Labels.append('Normal')

for file in os.listdir(infectedPath):
    image = Image.open(os.path.join(infectedPath, file)).resize((224, 224))
    Images.append(image)
    if 'bacteria' in file:
        Labels.append('Infected_Bacterial')
        infectedAmountBacterial += 1
    if 'virus' in file:
        Labels.append('Infected_Viral')
        infectedAmountViral += 1

print("Normal Images: ", normalAmount)
print("Infected Images: ", infectedAmountBacterial + infectedAmountViral)
print("Infected Bacterial Images: ", infectedAmountBacterial)
print("Infected Viral Images: ", infectedAmountViral)



```
