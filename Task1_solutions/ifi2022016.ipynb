```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import os
import PIL as Image
import cv2
```

```python
images = []
label = []
infected_path = '/content/OpenCodeData/OpenCodeData/INFECTED'
normal_path = '/content/OpenCodeData/OpenCodeData/NORMAL'
```

```python

for img in os.listdir(infected_path):
  file_path = os.path.join(infected_path, img)
  if file_path.endswith(('.jpg', '.jpeg', '.png')):
      image = cv2.imread(file_path, cv2.IMREAD_GRAYSCALE)
      if "virus" in img:
            images.append(image)
            label.append("virus")
      elif "bacteria" in img:
          images.append(image)
          label.append("bacteria")
```

```python
for img in os.listdir(normal_path):
  file_path = os.path.join(normal_path, img)
  if file_path.endswith(('.jpg', '.jpeg', '.png')):
      image = cv2.imread(file_path, cv2.IMREAD_GRAYSCALE)
      label.append("normal")
      images.append(image)
```

```python
inf_vir = label.count("virus")
inf_bact = label.count("bacteria")
normal = label.count("normal")
```
